```python
#Checking the Inter Quartile Range of Pregnancies
Q1= df['Pregnancies'].quantile(0.25)
Q3= df['Pregnancies'].quantile(0.75)

IQR= Q3 - Q1

Lower_whisker = Q1 - 1.5 * IQR
Upper_whisker = Q3 + 1.5 * IQR

print(Lower_whisker)
print(Upper_whisker)

print(Q1)
print(Q3)
print(IQR)
```

```python
#Max values
df.loc[df['Pregnancies'].idxmax()]
```

```python
#Checking the outliers of Pregnancies
outliers = df[(df['Pregnancies'] < Lower_whisker) | (df['Pregnancies'] > Upper_whisker)]
```
